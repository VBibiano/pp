```python
#%%

df = pd.read_csv("boston_housing.csv",index_col=0 )
df.describe()
```

```python
#%%

bins = [0, 35, 70, max(df["AGE"])]
labels = ['A0', 'A1', 'A2']
df['AGE_Groups'] = pd.cut(df['AGE'], bins, labels=labels, right=False)
df.boxplot("MEDV","AGE_Groups")
```
